```python
#연습문제 3.2.24
n = 3
for a1 in range(10):
    a = a1 + 1
    for b1 in range(10):
        b = b1 + 1
        for c1 in range(10):
            c = c1 + 1
            x = a ** n + b ** n
            y = c ** n
            if x - y ==0:
                print("페르마의 정리는 틀렸습니다.")
            else:
                print("페르마의 정리 증명 완료.")
```
