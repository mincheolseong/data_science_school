```python
#shift 연산 : 날짜 이동 등의 다양한 연산 가능
np.random.seed(0)
ts = pd.Series(np.random.randn(4), index = pd.date_range(
"2018 - 1 - 1", periods = 4, freq = "M"))
ts
```

```python
ts.shift(1) #shift 연산은 인덱스 그대로 데이터만 이동
```

```python
ts.shift(-1)
```

```python
ts.shift(1, freq = "M") #shift(freq) 연산은 데이터 그대로 인덱스만 이동
```

```python
ts.shift(1, freq="W")
```

```python
#resample 연산 : 시간 간격 재조정하는 리샘플링 가능
#upsampling, downsampling
ts = pd.Series(np.random.randn(100), index = pd.date_range(
"2018 - 1 - 1", periods = 100, freq = "D"))
ts.tail(20)
```

```python
ts.resample('W').mean() #'주'로 묶고 그룹 연산은 평균mean으로
```

```python
ts.resample('M').first() #'달'로 묶고 그룹연산은 first
```

```python
#시/분 단위는 가장 빠른값은 포함, 가장 늦은 값은 포함 X
ts = pd.Series(np.random.randn(60), index = pd.date_range(
"2018 - 1 - 1", periods = 60, freq = "T"))
ts
```

```python
ts.resample('10T').sum()
```

```python
#왼쪽이 아니라 오른쪽의 한계값을 포함하려면 closed="right"인수
ts.resample('10T', closed = "right").sum()
```

```python
#ohlc : 구간의 open, high, low, close 값을 구함
ts.resample('5T').ohlc()
```

```python
#In case of up-sampling, you should make data that does not actually exits
#foward filling, backward filling 방식
ts.resample('30s').ffill().head(20)
```

```python
ts.resample('30s').bfill().head(20)
```
