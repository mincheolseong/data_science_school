```python
#merge : 두 데이터 프레임의 공통 열 또는 인덱스 기준 병합
#이 때 기준이 되는 열, 행의 데이터를 키key라고 한다
df1 = pd.DataFrame({
    '고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름' : ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns = ['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
#merge로 df1, df2합칠 때 공통 열 고객번호 기준으로 데이터 찾아서 합친다
#양쪽에 모두 키가 존재하는 데이터만 보여줌!!
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how='outer') #outer join방식은 키 값이 한쪽에만 있어도
```

```python
pd.merge(df1,df2, how='left')
```

```python
pd.merge(df1,df2, how='right')
```

```python
df1 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns=['품종', '꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]},
    columns=['품종', '꽃잎너비'])
df2
```

```python
pd.merge(df1,df2) #키 값이 같은 데이터가 여러 개있으면 모든 경우의 수
```

```python
#이름이 같은 열 = key가 되는데 이름같아도 키가 되면 안되는 열 : on인수로 기준열 명시
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1 #데이터는 금액을 나타내는 데이터
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2 #데이터는 성별을 나타내는 데이터
```

```python
pd.merge(df1,df2, on = '고객명') # on인수로 기준열 강제로 설정
```

```python
#키가 되는 기준열의 이름이 다르면 left_on, right_on인수 사용
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적2': [4, 5, 6]})
df2
```

```python
#df1의 이름과 df2의 성명은 같은 데이터로 키가 되어야 함
pd.merge(df1, df2, left_on = '이름', right_on = '성명')
```

```python
#일반 데이터 열이 아닌 인덱스 기준으로 하려면 left_index, right_index 인수를 True로 사용
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547]})
df1
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape((6, 2)),
    index=[['부산', '부산', '서울', '서울', '서울', '서울'],
           [2000, 2005, 2000, 2005, 2010, 2015]],
    columns=['데이터1', '데이터2'])
df2
```

```python
pd.merge(df1,df2, left_on=['도시', '연도'], right_index = True)
```

```python
df1 = pd.DataFrame(
    [[1., 2.], [3., 4.], [5., 6.]],
    index=['a', 'c', 'e'],
    columns=['서울', '부산'])
df1
```

```python
df2 = pd.DataFrame(
    [[7., 8.], [9., 10.], [11., 12.], [13, 14]],
    index=['b', 'c', 'd', 'e'],
    columns=['대구', '광주'])
df2
```

```python
pd.merge(df1,df2,how = 'outer', left_index = True, right_index = True)
```

```python
#join method # you can also use 'join' method, instead of 'merge'
df1.join(df2, how = 'outer')
```

```python
#연습 문제 4.2.10
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547],
    '번호': [0,1,2,3,4],
    '밥': [5,4,3,2,1]})
df1
```

```python
#연습 문제 4.2.11
df1 = pd.DataFrame({
    '1월' : [10, 5],
    '2월' : [10, 6],
    '3월' : [10, 7],
    '4월' : [10, 8],
    '5월' : [11, 5],
    '6월' : [12, 5]},
    index=['매출', '비용'],
    columns=['1월', '2월','3월','4월','5월','6월'])
df1.loc["이익",:] = df1.apply(lambda x : x.max() - x.min())
df1
```

```python
df2 = pd.DataFrame({
    '7월' : [20, 8],
    '8월' : [20, 3],
    '9월' : [20, 7],
    '10월' : [20, 8],
    '11월' : [21, 5],
    '12월' : [22, 5]},
    index=['매출', '비용'],
    columns=['7월', '8월','9월','10월','11월','12월'])
df2.loc["이익",:] = df2.apply(lambda x : x.max() - x.min())
df2
```

```python
df3 = pd.concat([df1,df2], axis = 1)
df3
```

```python
df3["총 실적"] = df3.sum(axis = 1)
df3
```
