```python
%matplotlib inline
#바 차트
import matplotlib as mpl
import matplotlib.pylab as plt
import matplotlib.font_manager as fm

path = 'C:\\Users\\82104\\Downloads\\NanumBarunGothic.ttf'
fontprop = fm.FontProperties(fname = path, size = 15)

y = [2, 3, 1]
x = np.arange(len(y)) #x = 012 0123 01
xlabel = ['a', 'b', 'c']
plt.title("Bar Chart")
plt.bar(x, y) #x데이터가 카테고리일때 바 차트 시각화 : bar 명령
plt.xticks(x, xlabel)
plt.yticks(sorted(y)) #sorted는 y를 오름차순 정리
plt.xlabel("가나다", fontproperties = fontprop)
plt.ylabel("빈도수", fontproperties = fontprop)
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people)) #y_pos = 0123
performance = 3 + 10 * np.random.rand(len(people)) #rand 4개
error = np.random.rand(len(people)) #error bar 추가하기

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha = 0.4)
plt.yticks(y_pos, people, fontproperties = fontprop)
plt.xlabel("x라벨", fontproperties = fontprop)
plt.show()
```

```python
#스템 플롯 : 바 차트와 비슷하지만 폭width가 없는 stem plot
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")
plt.stem(x, np.cos(x), '-.')
plt.show()
```

```python
#파이 차트 : 원의 형태률 유지키위해 plt.axis('equal') 사용
labels = ['frog', 'pig', 'dog', 'log']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.2, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
       autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.show()
```

```python
#히스토그램 :
# hist 명령은 bins인수로 데이터 집계 구간 정보 받음
np.random.seed(0)
x = np.random.randn(1000)
plt.title("Histogram")
arrays, bins, patches = plt.hist(x, bins = 10)
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12, 3),
                         subplot_kw={'xticks': [], 'yticks': []})
axes[0].set_title("plt.cm.Blues")
axes[0].imshow(X, interpolation='nearest', cmap=plt.cm.Blues)
axes[1].set_title("plt.cm.Blues_r")
axes[1].imshow(X, interpolation='nearest', cmap=plt.cm.Blues_r)
axes[2].set_title("plt.BrBG")
axes[2].imshow(X, interpolation='nearest', cmap=plt.cm.BrBG)
axes[3].set_title("plt.BrBG_r")
axes[3].imshow(X, interpolation='nearest', cmap='BrBG_r')
plt.show()
```

```python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
          'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
          'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']
fig, axes = plt.subplots(3, 6, figsize = (12, 6), 
                         subplot_kw={'xticks' : [], 'yticks' : []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

```python
#컨투어contour플롯
#입력변수x,y2개, 출력변수z1개
#입력변수x,y는 'meshgrid'로 그리드 포인트 행렬로 바꿔서 입력
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
#3D 서피스surface 명령 by 'Axes3D'
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
