```python
#다중 인덱스 #행이나 열에 여러 계층
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5,4), 2),
                  columns = [["A", "A", "B", "B"],
                             ["C1", "C2", "C1", "C2"]])
#coulumns 인수를 다음과 같이 설정 -> multi-index
df3
```

```python
df3.columns.names = ["Cidx1", "Cidx2"]
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4), 2),
                  columns = [["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                  index = [["M", "M","M","F","F","F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

```python
#행 인덱스과 열 인덱스 교환
#stack() : 열 -> 행
#unstack() : 행 -> 열
df4.stack("Cidx1")
```

```python
df4.unstack("Ridx2")
```

```python
df4.unstack(0) # (0), ("Ridx1") 결과 같다
```

```python
#다중 인덱스가 있는 경우의 인덱싱
#다중 인덱스일 경우 인덱스는 하나의 라벨, 숫자가 아니라 ()의 튜플이 되어야
df3
```

```python
df3[("B","C1")]
```

```python
df3.loc[0,("B","C1")] #2차원 인덱싱
```

```python
df3.iloc[0,2] #iloc 인덱서의 경우 튜플()형태의 다중인덱스 사용안됨
```

```python
df3["A"]
```

```python
df4.loc[("M","id_1"),("A","C")]
```

```python
df4.loc[:,("A","C")]
```

```python
df4
```

```python
df4.loc[("ALL","ALL"), :] = df4.sum()
df4
```

```python
#다중 인덱스의 인덱스 순서 교환 : swaplevel
df5 = df4.swaplevel("Ridx1", "Ridx2")
df5
```

```python
df6 = df4.swaplevel("Cidx1","Cidx2", 1)
df6
```

```python
#다중 인덱스가 있는 경우의 정렬
df5
```

```python
df5.sort_index(level = 0)
```

```python
df6
```

```python
df6.sort_index(axis = 1, level = 0)
```
