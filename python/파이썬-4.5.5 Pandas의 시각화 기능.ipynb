```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
iris = sns.load_dataset("iris")    #  
titanic = sns.load_dataset("titanic")    #  

iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("  ")
plt.xlabel("Data")
plt.ylabel(" ")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris  Bar Plot")
plt.xlabel("Data")
plt.ylabel(" Feature ")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.title("  Feature ")
plt.xlabel("")
plt.ylabel("")
plt.ylim(0, 8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title(" Feature   ")
plt.xlabel("Feature")
plt.ylabel("")
plt.show()
```

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("   ")
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.title(" Feature   Histogram")
plt.xlabel(" ")
plt.show()
```

```python
iris.plot.kde()
plt.title(" Feature    Kernel Density Plot")
plt.xlabel(" ")
plt.show()
```

```python
iris.plot.box()
plt.title(" Feature    Box Plot")
plt.xlabel("Feature")
plt.ylabel(" ")
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=3, h_pad=1)
plt.title(" Feature     Box Plot")
plt.show()
```
